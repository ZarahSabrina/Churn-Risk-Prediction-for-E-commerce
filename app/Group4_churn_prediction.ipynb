```python
import pandas as pd

# Salin data awal
df_feature = final_dataset_clean.copy()

# Konversi tanggal ke datetime
df_feature["order_purchase_timestamp"] = pd.to_datetime(df_feature["order_purchase_timestamp"])
last_date = df_feature["order_purchase_timestamp"].max()

# Hitung tanggal transaksi terakhir per pelanggan
last_transaction = (
    df_feature.groupby("customer_unique_id")["order_purchase_timestamp"]
    .max()
    .reset_index(name="last_order_date")
)

# Hitung tanggal transaksi pertama per pelanggan
first_transaction = (
    df_feature.groupby("customer_unique_id")["order_purchase_timestamp"]
    .min()
    .reset_index(name="first_order_date")
)

# Hitung durasi aktif (dalam hari) dari pembelian pertama hingga last_date
customer_lifecycle = (
    first_transaction
    .merge(last_transaction, on="customer_unique_id")
)

customer_lifecycle["retention_days"] = (
    (last_date - customer_lifecycle["first_order_date"]).dt.days
)

# Tetapkan churn: tidak transaksi dalam 6 bulan terakhir
churn_threshold = last_date - pd.DateOffset(months=6)
customer_lifecycle["churn"] = (customer_lifecycle["last_order_date"] < churn_threshold).astype(int)

# Agregasi fitur numerik dan payment_type
agg_feature = df_feature.groupby("customer_unique_id").agg(
    total_payment_value=("payment_value", "sum"),
    mean_price=("price", "mean"),
    avg_review_score=("review_score", "mean"),
    payment_type=("payment_type", lambda x: x.mode().iloc[0] if not x.mode().empty else None)
).reset_index()

# Ambil modus kategori produk per pelanggan
product_category_name = df_feature.groupby("customer_unique_id")["product_category_name"].agg(
    lambda x: x.mode().iloc[0] if not x.mode().empty else None
).reset_index()

# Ambil modus customer_state per pelanggan
customer_state = df_feature.groupby("customer_unique_id")["customer_state"].agg(
    lambda x: x.mode().iloc[0] if not x.mode().empty else None
).reset_index()

# Simpan semua order_purchase_timestamp sebagai list per pelanggan
timestamp_list = df_feature.groupby("customer_unique_id")["order_purchase_timestamp"].agg(list).reset_index()

# Gabungkan semua fitur
df_feature = (
    agg_feature
    .merge(product_category_name, on="customer_unique_id", how="left")
    .merge(customer_lifecycle[["customer_unique_id", "retention_days", "churn", "last_order_date"]], on="customer_unique_id", how="left")
    .merge(customer_state, on="customer_unique_id", how="left")
    .merge(timestamp_list, on="customer_unique_id", how="left")  # Tambahkan daftar timestamp
)

# Buat kategori spending berdasarkan total_payment_value
bins_quartiles = [0, 61, 189, float("inf")]
labels_quartiles = ["Low Spend", "Medium Spend", "High Spend"]
df_feature["spending_category"] = pd.cut(
    df_feature["total_payment_value"],
    bins=bins_quartiles,
    labels=labels_quartiles
)
```

```python
import pandas as pd
import scipy.stats as stats
from scipy.stats import chi2_contingency

# Variabel kategori yang akan diuji
categorical_vars = ["customer_region","spending_category"]

# Simpan hasil uji statistik
anova_results = {}
chi_square_results = {}

# ** Uji ANOVA: Menganalisis rata-rata churn antar kategori**
for cat_var in categorical_vars:
    # Hitung jumlah kategori unik (tanpa NaN)
    unique_levels = df_feature[cat_var].dropna().nunique()
    
    if unique_levels > 1:  # Pastikan ada lebih dari satu kategori
        groups = [df_feature[df_feature[cat_var] == level]["churn"].dropna() 
                  for level in df_feature[cat_var].dropna().unique()]
        
        # Pastikan tidak ada kategori yang kosong atau memiliki satu nilai saja
        if len(groups) > 1 and all(len(g) > 1 for g in groups) and all(g.var() > 0 for g in groups):
            f_stat, p_value = stats.f_oneway(*groups)  # Uji ANOVA
            anova_results[cat_var] = {"F-statistic": f_stat, "P-value": p_value}

# ** Uji Chi-Square: Hubungan antara churn dengan kategori**
for cat_var in categorical_vars:
    unique_levels = df_feature[cat_var].dropna().nunique()

    if unique_levels > 1:  # Pastikan ada lebih dari satu kategori
        contingency_table = pd.crosstab(df_feature[cat_var].dropna(), df_feature["churn"])
        
        # Pastikan tabel kontingensi valid (tidak ada baris yang semua 0)
        if (contingency_table.values > 0).sum() > 1:
            chi2, p, dof, expected = chi2_contingency(contingency_table)
            chi_square_results[cat_var] = {"Chi2": chi2, "P-value": p}

# ** Tampilkan Hasil Uji ANOVA**
print("===== ANOVA Test Results (Churn vs Categorical Variables) =====")
if anova_results:
    for key, value in anova_results.items():
        significance = "There is a significant difference" if value['P-value'] <= 0.05 else "There is no significant difference"
        print(f"{key}: F={value['F-statistic']:.4f}, P={value['P-value']:.2e}  {significance}")
else:
    print("Tidak ada variabel yang memenuhi syarat untuk uji ANOVA.")

# ** Tampilkan Hasil Uji Chi-Square**
print("\n===== Chi-Square Test Results (Churn vs Categorical Variables) =====")
if chi_square_results:
    for key, value in chi_square_results.items():
        significance = "There is a significant dependence" if value['P-value'] <= 0.05 else "There is no significant difference"
        print(f"{key}: Chi2={value['Chi2']:.4f}, P={value['P-value']:.2e}  {significance}")
else:
    print("Tidak ada variabel yang memenuhi syarat untuk uji Chi-Square.")
```

```python
from scipy.stats import ttest_ind

numerical_vars = ["mean_price", "total_payment_value", "avg_review_score"]  # Sesuaikan dengan dataset

t_test_results = {}

for num_var in numerical_vars:
    churn_group = df_feature[df_feature["churn"] == 1][num_var].dropna()
    no_churn_group = df_feature[df_feature["churn"] == 0][num_var].dropna()
    
    if len(churn_group) > 1 and len(no_churn_group) > 1:
        stat, p_value = ttest_ind(churn_group, no_churn_group, equal_var=False)  # Uji t-test
        significance = "There is a significant difference" if p_value <= 0.05 else "There is a no significant difference"
        t_test_results[num_var] = {"T-statistic": stat, "P-value": p_value, "Interpretasi": significance}

# ** Tampilkan Hasil Uji t-Test**
print("\n===== T-Test Results (Numerical vs Churn) =====")
for key, value in t_test_results.items():
    print(f"{key}: T={value['T-statistic']:.4f}, P={value['P-value']:.2e}  {value['Interpretasi']}")
```

```python
import pandas as pd
import numpy as np

# Fungsi untuk menghitung WOE & IV
def calculate_woe_iv(df, feature, target):
    """
    Menghitung Weight of Evidence (WOE) dan Information Value (IV) untuk variabel kategori.
    
    Parameters:
        df (pd.DataFrame): DataFrame yang berisi fitur dan target.
        feature (str): Nama kolom fitur kategori.
        target (str): Nama kolom target biner (0/1).
    
    Returns:
        pd.DataFrame: Tabel WOE & IV untuk setiap kategori dalam fitur.
        float: Nilai IV keseluruhan untuk fitur tersebut.
    """
    # Buat tabel distribusi
    df_woe = df_feature.groupby(feature)[target].agg(['count', 'sum'])
    
    # Rename kolom
    df_woe.columns = ['Total', 'Churn']
    df_woe['Non-Churn'] = df_woe['Total'] - df_woe['Churn']
    
    # Hitung proporsi
    df_woe['% Churn'] = df_woe['Churn'] / df_woe['Churn'].sum()
    df_woe['% Non-Churn'] = df_woe['Non-Churn'] / df_woe['Non-Churn'].sum()
    
    # Hitung WOE (hindari log(0) dengan +0.0001)
    df_woe['WOE'] = np.log((df_woe['% Churn'] + 0.0001) / (df_woe['% Non-Churn'] + 0.0001))
    
    # Hitung IV
    df_woe['IV'] = (df_woe['% Churn'] - df_woe['% Non-Churn']) * df_woe['WOE']
    iv_value = df_woe['IV'].sum()
    
    return df_woe[['Total', 'Churn', 'Non-Churn', '% Churn', '% Non-Churn', 'WOE', 'IV']], iv_value

# ** Uji WOE & IV untuk Setiap Variabel Kategori**
categorical_vars = ["customer_region","spending_category","payment_type","product_category_group"]
woe_results = {}
iv_scores = {}

for cat_var in categorical_vars:
    woe_table, iv_score = calculate_woe_iv(df_feature, cat_var, "churn")
    woe_results[cat_var] = woe_table
    iv_scores[cat_var] = iv_score

# ** Tampilkan hasil IV untuk setiap fitur**
print("\n===== Information Value (IV) Scores =====")
for feature, iv in iv_scores.items():
    strength = "Low" if iv < 0.02 else "Medium" if iv < 0.1 else "High" if iv < 0.3 else "Sangat Kuat"
    print(f"{feature}: IV={iv:.4f}  {strength}")

# ** Tampilkan hasil WOE untuk fitur tertentu**
selected_feature = "product_category_group"  # Ganti sesuai kebutuhan
if selected_feature in woe_results:
    print(f"\n===== WOE Table for {selected_feature} =====")
    print(woe_results[selected_feature])
```

```python
from sklearn.metrics import  recall_score, roc_auc_score

def evaluate_model(model, X_train, y_train, X_test, y_test):
    print(" Model Evaluation")

    #  Prediksi pada training set
    y_train_pred = model.predict(X_train)
    y_train_proba = model.predict_proba(X_train)[:, 1]

    #  Prediksi pada testing set
    y_test_pred = model.predict(X_test)
    y_test_proba = model.predict_proba(X_test)[:, 1]

    #  Evaluasi Training
    print("\n Training Set:")
    print(f"Recall   : {recall_score(y_train, y_train_pred):.4f}")
    print(f"ROC AUC  : {roc_auc_score(y_train, y_train_proba):.4f}")

    #  Evaluasi Testing
    print("\n Testing Set:")
    print(f"Recall   : {recall_score(y_test, y_test_pred):.4f}")
    print(f"ROC AUC  : {roc_auc_score(y_test, y_test_proba):.4f}")
```

#### MONEY EVALUATION

```python
# Pastikan kolom 'churn' ada dan bertipe biner (0 = not churn, 1 = churn)
churn_df = df_feature[df_feature['churn'] == 1]
non_churn_df = df_feature[df_feature['churn'] == 0]

# --- SAMPLING DATA ---
tp_customers = churn_df.sample(n=scaled_tp, random_state=42)
fp_customers = non_churn_df.sample(n=scaled_fp, random_state=42)
remaining_churn_df = churn_df.drop(tp_customers.index)
fn_customers = remaining_churn_df.sample(n=scaled_fn, random_state=42)

# --- REVENUE IMPACT ANALYSIS ---

# Revenue masing-masing kelompok
tp_total_revenue = tp_customers["total_payment_value"].sum()
fp_total_revenue = fp_customers["total_payment_value"].sum()
fn_total_revenue = fn_customers["total_payment_value"].sum()
total_churned_revenue = churn_df["total_payment_value"].sum()

# Asumsi
discount_rate = 0.07
marketing_cost = 1
tp_recovery_rate = 0.30
fp_loyal_rate = 0.15

# TP impact
tp_discount = tp_total_revenue * discount_rate
tp_marketing = scaled_tp * marketing_cost
tp_net_revenue = tp_total_revenue - tp_discount - tp_marketing
tp_retained_revenue = tp_net_revenue * tp_recovery_rate

# FP impact
fp_marketing = scaled_fp * marketing_cost
fp_cost = fp_marketing
fp_loyal_revenue = fp_total_revenue * fp_loyal_rate
fp_net_gain = fp_loyal_revenue - fp_cost

# FN impact
fn_lost_revenue = fn_total_revenue

# Total Cost
total_discount = tp_discount 
total_marketing_cost = tp_marketing + fp_marketing
total_cost = total_discount + total_marketing_cost

# Total Net Gain
total_net_gain = tp_retained_revenue + fp_net_gain
total_net_gain_pct = (total_net_gain / total_churned_revenue) * 100
fn_loss_pct = (fn_lost_revenue / total_churned_revenue) * 100

# Hitung projected retained revenue
projected_retained_revenue = total_churned_revenue + total_net_gain - total_cost


# Ringkasan Revenue
revenue_summary = pd.DataFrame({
    "Money Evaluation": [
        "Total Churned Revenue",
        "TP Retained Revenue (30%)",
        "FP Net Gain (15% loyal)",
        "Total Net Gain (TP + FP)",
        "Net Gain % of Churned Revenue",
        "FN Lost Revenue",
        "FN Loss % of Churned Revenue",
        "Total Discount Given (TP)",
        "Total Marketing Cost (TP + FP)",
        "Total Cost (Discount + Marketing)",
        "Projected Retained Revenue"
    ],
    "Value": [
        f"${total_churned_revenue:,.2f}",
        f"${tp_retained_revenue:,.2f}",
        f"${fp_net_gain:,.2f}",
        f"${total_net_gain:,.2f}",
        f"{total_net_gain_pct:.2f}%",
        f"${fn_lost_revenue:,.2f}",
        f"{fn_loss_pct:.2f}%",
        f"${total_discount:,.2f}",
        f"${total_marketing_cost:,.2f}",
        f"${total_cost:,.2f}",
        f"${projected_retained_revenue:,.2f}"
    ]
})

revenue_summary
```
